```python
# Importing libraries :
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset to pandas dataframe :
cc_data = pd.read_csv('/content/sample_data/creditcard.csv')
```

```python
# Top 5 entries :
cc_data.head()
```

```python
# Dataset information :
cc_data.info()
```

```python
# Checking number of missing values in each column :
cc_data.isnull().sum()
```

```python
# The last row has null values so we can drop the last row :
cc_data.drop(49609, inplace=True)
```

```python
# Distribution of legit transaction and fraudulent transaction :
cc_data['Class'].value_counts()
```

<b> This dataset is highly unbalanced : </b>
</br>
0 ---> Normal Transaction (large) </br>
1 ---> Fraud Transaction (small)

```python
# Separating the data for analysis :
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data :
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for tboth transaction :
cc_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=148)
```

```python
# Concatenating two dataframes :
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

<b> Splitting data into Features and Targets :

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

<b> Splitting into Train and Test data :</b>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, stratify=Y, random_state=2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

<b> Model Training : Logistic Regrression (Binary Classification) : </b> 

```python
model = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
# Training the Logistic regression model with training data :
model.fit(X_train, Y_train)
```

```python
# Accuracy on training data :
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data :' ,training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data :' , test_data_accuracy)
```
